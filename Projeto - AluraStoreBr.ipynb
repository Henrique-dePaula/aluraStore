### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Carregar os arquivos CSV por url

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#0. Unificar os dados para análise comparativa

```python
import matplotlib.pyplot as plt


# Adicionar identificação das lojas
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Unificar os dados em um único DataFrame
dados = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)
dados.head()
```

#1. Análise do faturamento


```python
import matplotlib.pyplot as plt

# Faturamento total por loja
faturamento = dados.groupby('Loja')['Preço'].sum()

# Reordenar para a ordem desejada: Loja 1, Loja 2, Loja 3, Loja 4
ordem_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamento = faturamento.reindex(ordem_lojas)

# Plotar gráfico de barras
faturamento.plot(kind='bar', title='Faturamento Total por Loja', ylabel='Faturamento (R$)', xlabel='Loja')
plt.tight_layout()
plt.show()

# Exibir a tabela
faturamento
```

### 🔍 Análise

- A **Loja 1** foi a que mais faturou, com mais de **R\$ 1,5 milhão** em vendas.
- A **Loja 4** foi a que menos faturou, com cerca de **R\$ 1,38 milhão**.

---

### 🏁 Conclusão

A **Loja 4** é a menos eficiente em faturamento, sendo a loja com o menor volume financeiro total de vendas entre as quatro analisadas.

# 2. Vendas por Categoria


```python
import matplotlib.pyplot as plt
import pandas as pd

def resumo_categorias_loja(dados, nome_loja):
    dados_loja = dados[dados['Loja'] == nome_loja]

    # Contar categorias e ordenar da maior para a menor
    categorias = dados_loja['Categoria do Produto'].value_counts().sort_values(ascending=False)

    # Plotar gráfico pizza
    categorias.plot(kind='pie', autopct='%1.1f%%', title=f'Categorias mais vendidas - {nome_loja}', ylabel='')
    plt.tight_layout()
    plt.show()

    # Criar DataFrame para exibição
    tabela_categorias = pd.DataFrame({
        'Categoria do Produto': categorias.index,
        'Quantidade de Vendas': categorias.values
    })

    # Exibir título e tabela formatada no notebook
    print(f"Categorias vendidas na {nome_loja} (da maior para a menor):")
    display(tabela_categorias)  # display() funciona em notebooks para formatar bonito

    # Definir maior e menor quantidade vendida
    max_vendas = categorias.max()
    min_vendas = categorias.min()

    # Listar todas categorias empatadas em maior e menor vendas
    categorias_mais_vendidas = categorias[categorias == max_vendas].index.tolist()
    categorias_menos_vendidas = categorias[categorias == min_vendas].index.tolist()

    if len(categorias_mais_vendidas) > 1:
        print(f"\nCategoria(s) mais vendida(s) da {nome_loja}: {', '.join(categorias_mais_vendidas)} ({max_vendas} vendas cada)")
    else:
        print(f"\nCategoria mais vendida da {nome_loja}: {categorias_mais_vendidas[0]} ({max_vendas} vendas)")

    if len(categorias_menos_vendidas) > 1:
        print(f"Categoria(s) menos vendida(s) da {nome_loja}: {', '.join(categorias_menos_vendidas)} ({min_vendas} vendas cada)")
    else:
        print(f"Categoria menos vendida da {nome_loja}: {categorias_menos_vendidas[0]} ({min_vendas} vendas)")

    # Divisória visual para separar lojas
    print("\n" + "="*50 + "\n")

# Exemplo de uso para as 4 lojas:
for loja in ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']:
    resumo_categorias_loja(dados, loja)
```

# 📊 Análise das Categorias de Produtos Vendidas por Loja

## Visão Geral

Em todas as quatro lojas, a categoria **"moveis"** é consistentemente a mais vendida, com números que variam de 442 a 499 unidades vendidas, indicando que móveis são os produtos mais populares e provavelmente o principal motor de vendas em todas as lojas.

Por outro lado, as categorias menos vendidas apresentam mais variação entre as lojas:

- Loja 1 e Loja 2 têm **"utilidades domesticas"** como a categoria menos vendida, com 171 e 181 vendas, respectivamente.
- Loja 3 apresenta empate entre **"instrumentos musicais"** e **"utilidades domesticas"**, ambas com 177 vendas, indicando uma distribuição próxima para o menor interesse nessas categorias.
- Loja 4 tem **"instrumentos musicais"** como a categoria menos vendida, com 170 vendas.
---
## Observações

- A categoria **"moveis"** lidera confortavelmente em todas as lojas, demonstrando um forte apelo e possivelmente um estoque bem ajustado para essa categoria.
- **"utilidades domesticas"** e **"instrumentos musicais"** são categorias consistentemente menos populares, o que pode indicar oportunidade para estratégias de marketing específicas para essas categorias, ou reavaliação do estoque para evitar excesso de produtos.
- O empate na Loja 3 no menor volume de vendas entre duas categorias sugere que a atenção pode ser dividida entre ambas para melhorias.

# 4. Produtos Mais e Menos Vendidos

```python
def resumo_produtos_extremo_loja_com_categoria(dados, nome_loja):
    dados_loja = dados[dados['Loja'] == nome_loja]

    # Contagem de vendas por produto
    produtos = dados_loja['Produto'].value_counts()

    max_vendas = produtos.max()
    min_vendas = produtos.min()

    produtos_mais_vendidos = produtos[produtos == max_vendas].index.tolist()
    produtos_menos_vendidos = produtos[produtos == min_vendas].index.tolist()

    print(f"\nLoja: {nome_loja}")

    # Produtos mais vendidos com categoria
    if len(produtos_mais_vendidos) > 1:
        mais_vendidos_info = []
        for prod in produtos_mais_vendidos:
            categoria = dados_loja.loc[dados_loja['Produto'] == prod, 'Categoria do Produto'].mode()[0]
            mais_vendidos_info.append(f"{prod} ({categoria})")
        mais_vendidos_str = ', '.join(mais_vendidos_info)
        print(f"Produto(s) mais vendido(s): {mais_vendidos_str} - {max_vendas} vendas cada")
    else:
        categoria = dados_loja.loc[dados_loja['Produto'] == produtos_mais_vendidos[0], 'Categoria do Produto'].mode()[0]
        print(f"Produto mais vendido: {produtos_mais_vendidos[0]} ({categoria}) - {max_vendas} vendas")

    # Produtos menos vendidos com categoria
    if len(produtos_menos_vendidos) > 1:
        menos_vendidos_info = []
        for prod in produtos_menos_vendidos:
            categoria = dados_loja.loc[dados_loja['Produto'] == prod, 'Categoria do Produto'].mode()[0]
            menos_vendidos_info.append(f"{prod} ({categoria})")
        menos_vendidos_str = ', '.join(menos_vendidos_info)
        print(f"Produto(s) menos vendido(s): {menos_vendidos_str} - {min_vendas} vendas cada")
    else:
        categoria = dados_loja.loc[dados_loja['Produto'] == produtos_menos_vendidos[0], 'Categoria do Produto'].mode()[0]
        print(f"Produto menos vendido: {produtos_menos_vendidos[0]} ({categoria}) - {min_vendas} vendas")

# Rodar para as 4 lojas
for loja in ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']:
    resumo_produtos_extremo_loja_com_categoria(dados, loja)
```

### 🔍 Análise dos Produtos Mais e Menos Vendidos por Loja

- **Loja 1** apresenta uma diversidade entre os produtos mais vendidos, com itens das categorias eletrodomésticos, eletrônicos e móveis alcançando o topo das vendas, todos com 60 unidades. Isso indica uma boa distribuição das preferências dos clientes entre diferentes tipos de produtos. Já os produtos menos vendidos são ambos eletrônicos, sugerindo talvez menor apelo em alguns eletrônicos específicos.

- **Loja 2** tem um destaque claro para o produto “Iniciando em programação” (categoria livros), que lidera com 65 vendas. O produto menos vendido está na categoria brinquedos, indicando que esta loja pode ter menor foco ou demanda para brinquedos.

- **Loja 3** mostra forte desempenho em móveis, com “Kit banquetas” liderando as vendas. O produto menos vendido é um brinquedo, o que pode indicar menor relevância dessa categoria na loja.

- **Loja 4** também destaca os móveis como produtos mais vendidos, especialmente a “Cama box”. Por outro lado, o produto menos vendido é um instrumento musical, o que pode sugerir menor popularidade ou estoque reduzido dessa categoria na loja.

---

### 🏁 Conclusão Geral

A análise indica que as lojas 1, 3 e 4 têm forte apelo para móveis, enquanto a loja 2 se destaca por vendas de livros. Produtos menos vendidos tendem a estar mais concentrados em categorias específicas, como eletrônicos ou brinquedos, dependendo da loja. Para melhorar a eficiência, pode ser interessante cada loja avaliar estratégias para aumentar a venda das categorias menos populares ou ajustar o mix de produtos para o perfil dos seus clientes.
